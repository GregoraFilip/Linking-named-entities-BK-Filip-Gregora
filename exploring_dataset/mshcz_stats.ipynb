```python
import pandas as pd
import xml.etree.ElementTree as elt
```

```python
mshcz = elt.parse('../databaze/MeSH2023_Marc21_Alma.xml').getroot()
```

```python
print(len(mshcz))
```

```python
n_synonyms = 0
n_names = 0
n_engl = 0
n_description = 0

for child in mshcz:
    for subchild in child.iter("{http://www.loc.gov/MARC21/slim}datafield"):
        tag = int(subchild.attrib["tag"])
        if tag == 150 or tag == 151 or tag == 155:
            n_names += 1
        if tag == 450 or tag == 451 or tag == 455:
            n_synonyms += 1
        if tag == 680:
            n_description += 1
        if tag == 750 or tag == 751 or tag == 755:
            n_engl += 1

print("Names:", n_names)
print("Synonyms:", n_synonyms)
print("English Names:", n_engl)
print("Descriptions:", n_description)
```

```python
n_cathegories = {}
for c, n in [("A", "Anatomie"),
             ("B", "Organismy"),
             ("C", "Nemoci"),
             ("D", "Chemikálie a léčiva"),
             ("E", "Analytické, diagnostické a terapeutické techniky a přístroje"),
             ("F", "Psychiatrie a psychologie"),
             ("G", "Jevy a procesy"),
             ("H", "Obory a povolání"),
             ("I", "Antropologie, vzdělávání, sociologie a sociální jevy"),
             ("J", "Technologie, průmysl a zemědělství"),
             ("K", "Humanitní vědy"),
             ("L", "Informační vědy"),
             ("M", "Lidé"),
             ("N", "Zdravotní péče"),
             ("V", "Publikační charakteristiky"),
             ("X", "Zastaralé"),
             ("Y", "Podhesla (kvalifikátory)"),
             ("Z", "Geografická místa"),
            ]:
    n_cathegories[c] = (0, 0, n)

for child in mshcz:
    try:
        n_synonyms = 0
        for subchild in child.iter("{http://www.loc.gov/MARC21/slim}datafield"):
            tag = int(subchild.attrib["tag"])
            if tag == 450 or tag == 451 or tag == 455:
                n_synonyms += 1
                
        d = [i for i in child.findall("{http://www.loc.gov/MARC21/slim}datafield") if i.attrib["tag"] == "072" ]
        for c in set([e[0].text[0] for e in d if e[0].text is not None]):
            n_cathegories[c] = (n_cathegories[c][0] + 1, n_cathegories[c][1] + n_synonyms, n_cathegories[c][2])
    except IndexError:
        pass
    except KeyError:
        print([i for i in child.findall("{http://www.loc.gov/MARC21/slim}controlfield") if i.attrib["tag"] == "001" ][0].text)


code = []
n_records = []
n_syn = []
name = []
for c, l in n_cathegories.items():
    code.append(c)
    n_records.append(l[0])
    n_syn.append(l[1])
    name.append(l[2])

occurence = pd.DataFrame({"name":name, "records":n_records, "synonyms":n_syn}, index=code)
```

```python
occurence.head(3)
```

```python
occurence["total"] = occurence.apply(lambda x: x.records + x.synonyms, axis=1)

print("In total records:", occurence["records"].sum())
print("In total synonyms:", occurence["synonyms"].sum())
print("In total synonyms and records:", occurence["total"].sum())
```

```python
occurence.plot( y="records", kind="bar", title="Number of records per cathegory", legend=False, ylabel="Number of evidence", xlabel="MeSH Code")
```

```python
occurence.plot( y="total", kind="bar", title="Number of records per cathegory", legend=False, ylabel="Number of evidence", xlabel="MeSH Code")
```

```python
occurence.drop(["synonyms"], axis=1).to_csv("../csv_output/mshcz_stats.csv")
```
